```python
# Tahun-tahun yang akan diproses (hanya sampai 2017)
years = range(2013, 2018)

# Buat kolom final berdasarkan nilai asli atau estimasi
for year in years:
    actual_col = f'generation_gwh_{year}'
    est_col = f'estimated_generation_gwh_{year}'
    final_col = f'final_generation_gwh_{year}'
    
    # Isi nilai dari actual, jika NaN ambil dari estimasi
    df[final_col] = df[actual_col].fillna(df[est_col])

# Contoh: tampilkan 5 baris pertama dari final kolom
print(df[[f'final_generation_gwh_{year}' for year in years]].head())
```

```python
# Daftar kolom yang akan dihapus
drop_cols = [
    f'generation_gwh_{year}' for year in range(2013, 2018)
] + [
    f'estimated_generation_gwh_{year}' for year in range(2013, 2018)
] + [
    f'estimated_generation_note_{year}' for year in range(2013, 2018)
]

# Hapus kolom-kolom tersebut
df.drop(columns=drop_cols, inplace=True)

# Cek sisa kolom
print(df.columns)
```

```python
df_melted = df.melt(
    id_vars=[
        "country", "country_long", "name", "gppd_idnr", "capacity_mw", "latitude",
        "longitude", "primary_fuel", "commissioning_year", "source"
    ],
    value_vars=[
        "final_generation_gwh_2013", "final_generation_gwh_2014",
        "final_generation_gwh_2015", "final_generation_gwh_2016",
        "final_generation_gwh_2017"
    ],
    var_name="Tahun",
    value_name="final_generation_gwh"
)

# Ubah Tahun dari string menjadi angka tahun saja
df_melted["Tahun"] = df_melted["Tahun"].str.extract(r"(\d{4})").astype(int)
```

```python
import atoti as tt
```

```python
# Mulai session Atoti
session = tt.Session.start()

# Membuat store dari df_melted
store = session.read_pandas(
    df_melted,
    keys=["gppd_idnr", "Tahun"],  # Primary key gabungan
    table_name="Pembangkit"
)

# Membuat cube
cube = session.create_cube(store, "PembangkitCube")

# Alias untuk mempermudah akses
lvl = cube.levels
msr = cube.measures

# Hierarki
cube.hierarchies["Lokasi"] = [store["country"], store["country_long"]]
cube.hierarchies["Tahun"] = [store["Tahun"]]
cube.hierarchies["primary_fuel"] = [store["primary_fuel"]]
# Jika kolom owner dihapus, hierarki ini bisa di-skip atau diganti yang lain

# Measures
msr["Total Kapasitas (MW)"] = tt.agg.sum(store["capacity_mw"])
msr["Total Energi (GWh)"] = tt.agg.sum(store["final_generation_gwh"])
msr["Rata-rata Kapasitas (MW)"] = tt.agg.mean(store["capacity_mw"])
msr["Kapasitas per GWh"] = msr["Total Kapasitas (MW)"] / msr["Total Energi (GWh)"]
msr["Jumlah Pembangkit"] = tt.agg.count_distinct(store["gppd_idnr"])
msr["Jumlah Pembangkit"] = tt.agg.count_distinct(store["gppd_idnr"])
```

```python
print(list(cube.levels.keys()))
```

### MEMBUAT DAN MENJAWAB PERTANYAAN MENGGUNAKAN CUBE.QUERY 

TOTAL ENERGI PER TAHUN

```python
result = cube.query(
    msr["Total Energi (GWh)"],
    levels=[lvl["Tahun"]]
)
print("Total Energi (GWh) per Tahun:")
print(result)
```

RATA-RATA KAPASITAS dan TOTAL KAPASITAS PER JENIS PEMBANGKIT

```python
result_capacity_per_fuel = cube.query(
    msr["Rata-rata Kapasitas (MW)"], msr["Total Kapasitas (MW)"],
    levels=[lvl["primary_fuel"]]
)

print("\nRata-rata dan Total Kapasitas per Jenis Pembangkit:")
print(result_capacity_per_fuel)
```

KAPASITAS PER LOKASI

```python
result_capacity_per_gwh_per_country = cube.query(
    msr["Kapasitas per GWh"],
    levels=[lvl[("Pembangkit", "Lokasi", "country")]]
)

print("\nKapasitas per GWh per Country:")
result_capacity_per_gwh_per_country
```

TOTAL ENERGI PER JENIS PEMBANGKIT

```python
result_energy_per_fuel = cube.query(
    msr["Total Energi (GWh)"],
    levels=[lvl[("Pembangkit","primary_fuel","primary_fuel")]]
)

result_energy_per_fuel
```

NEGARA DENGAN TOTAL ENERGI TERBESAR

```python
egara_Energi_Terbesar = cube.query(
    cube.measures["Total Energi (GWh)"],  # Pastikan nama measure yang sesuai
    levels=[cube.levels[("Pembangkit", "Lokasi", "country")]]
)

# Urutkan dari yang terbesar ke terkecil
Negara_Energi_Terbesar = Negara_Energi_Terbesar.sort_values(
    by=["Total Energi (GWh)"],  # Pastikan kolom sesuai dengan output query
    ascending=False
)

Negara_Energi_Terbesar
```

Bahan bakar paling banyak digunakan di seluruh negara

```python
Bahan_Bakar_Terpopuler = cube.query(
    msr["Total Kapasitas (MW)"],
    levels=[lvl["primary_fuel"]]
).sort_values(by=["Total Kapasitas (MW)"], ascending=False)

Bahan_Bakar_Terpopuler.head(5)  # Menampilkan 5 bahan bakar utama
```

Jenis bahan bakar yang paling banyak pembangkitnya di setiap negara

```python
Bahan_Bakar_Populer_Negara = cube.query(
    msr["Jumlah Pembangkit"],  # Gunakan Measure yang sudah terdaftar
    levels=[lvl["Pembangkit", "Lokasi", "country"], lvl[("primary_fuel")]]
).sort_values(by=["Jumlah Pembangkit"], ascending=False)

Bahan_Bakar_Populer_Negara.head(10)  # Menampilkan 10 kombinasi teratas
```
